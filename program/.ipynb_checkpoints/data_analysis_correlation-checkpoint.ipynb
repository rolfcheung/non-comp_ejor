```python
# Set notebook display
%matplotlib inline
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')

# import data analysis libraries
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read normalized data for SAW and WP method
normalized_data_path = os.path.join(os.path.abspath('..'), 'data', 'normalized_data.csv')

if os.path.exists(normalized_data_path):
    print('File paht:', normalized_data_path)
else:
    print('Check the data path!')

normalized_data = pd.read_csv(normalized_data_path, index_col = 0)
normalized_data.head()
```

```python
# read nest weights for SAW and WP method
nest_weight_path = os.path.join(os.path.abspath('..'), 'data', 'nest_weight.csv')

if os.path.exists(nest_weight_path):
    print('File paht:', nest_weight_path)
else:
    print('Check the data path!')

nest_weight = pd.read_csv(nest_weight_path, index_col = 0)
nest_weight
```

# SAW

```python
saw_matrix = normalized_data.copy()
saw_matrix['SAW'] = saw_matrix.dot(nest_weight.T)

saw_matrix['Rank'] = saw_matrix['SAW'].rank(ascending = False)

saw_matrix.head()
```

```python
saw_matrix.to_csv('../results/saw_wp/saw.csv')
```

# WP

```python
wp_matrix = normalized_data.copy()
```

```python
m, n = wp_matrix.shape

for row in range(m):
    for column in range(n):
        wp_matrix.iloc[row, column] = wp_matrix.iloc[row, column]** nest_weight.iloc[0, column]
        
wp_matrix.head()
```

```python
wp_matrix['WP'] = wp_matrix.sum(axis = 1)
wp_matrix['Rank'] = wp_matrix['WP'].rank(ascending = False)
wp_matrix.head()
```

```python
wp_matrix.to_csv('../results/saw_wp/wp.csv')
```
