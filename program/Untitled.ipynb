```python
# Set notebook display
%matplotlib inline
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')

# import data analysis libraries
import os
# import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# import seaborn as sns
```

```python
data_path = os.path.join(os.path.abspath('..'), 'data', 'normalized_data.csv')

if os.path.exists(data_path):
    print('File path:', data_path)
else:
    print('Check the path')
    
data = pd.read_csv(data_path, index_col = 0)
data.head()
```

```python
Uru_Har = data.loc[['Urumqi', 'Harbin']].T
Zun_Cha = data.loc[['Zunyi', 'Changsha']].T

ordered_Uru_Har = Uru_Har.sort_values(by = 'Urumqi')
ordered_Zun_Cha = Zun_Cha.sort_values(by = 'Zunyi')
```

```python
my_range = range(1, len(data.columns)+1)

font = {'family' : 'serif', 
        'color'  : 'k', 
        'weight' : 'normal',
        'size'   : 8
       }
```

```python
# fig = plt.figure(num =1, figsize = (20, 8))
# ax1 = plt.subplots(121)
fig = plt.figure(figsize = (20, 8))

ax1 = plt.subplot(121)



ax1.hlines(y = my_range,
           xmin = ordered_Uru_Har['Urumqi'],
           xmax = ordered_Uru_Har['Harbin'],
           colors = 'grey', alpha=0.4)
plt.yticks(my_range, ordered_Uru_Har.index)

plt.subplot(122)

plt.hlines(y = my_range,
           xmin = ordered_Zun_Cha['Zunyi'],
           xmax = ordered_Zun_Cha['Changsha'],
           colors = 'grey', alpha=0.4)
plt.yticks(my_range, ordered_Zun_Cha.index)

plt.tight_layout()

plt.show()
```
