```python
# import data analysis libraries
import os

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read data
raw_data_path = os.path.join(os.path.abspath('..'),
                             'data',
                             'rawdata.csv')

if os.path.exists(raw_data_path):
    print('File path:', raw_data_path)
else:
    print('Check the data path!')

raw_data = pd.read_csv(raw_data_path, index_col = 0)
raw_data.head()
```

```python
raw_data = raw_data.drop(['Latitude', 'Longitude', 'Location'], axis = 1)
```

```python
rank_data_path = os.path.join(os.path.abspath('..'),
                             'results',
                             'outranking_matrix',
                             'threshold_reference',
                             'threshold_reference.csv')

if os.path.exists(rank_data_path):
    print('File path: ', rank_data_path)
else:
    print('Check the path!')
    
rank_data = pd.read_csv(rank_data_path, index_col=0)
rank_data.head()
```

```python
indicator_rank_corr = pd.DataFrame(index = rank_data.columns,
                                  columns = raw_data.columns)
indicator_rank_corr.head()
```

```python
row, column = indicator_rank_corr.shape
```

```python
for row in range(row):
    for column in range(column):
        indicator_rank_corr.iloc[row, column] = rank_data.iloc[:, row].corr(raw_data.iloc[:, column])
        
indicator_rank_corr.head()
```
