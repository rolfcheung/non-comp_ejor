```python
# Set notebook display
%matplotlib inline
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')

# import data analysis libraries
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# correlation analysis:
data:
saw_rank, wp_rank, outranking rank (threshold reference)

```python
# read nest weights for SAW and WP method
corr_matrix_path = os.path.join(os.path.abspath('..'), 'results','corr_analysis',
                                'corr_analysis.csv')

if os.path.exists(corr_matrix_path):
    print('File paht:', corr_matrix_path)
else:
    print('Check the data path!')

corr_matrix = pd.read_csv(corr_matrix_path, index_col = 0)
corr_matrix.head()
```

```python
corr_matrix_data = corr_matrix.corr()
corr_matrix_data 
```

```python
from string import ascii_letters
```

```python
# generate a mask for the upper triangle
mask = np.zeros_like(corr_matrix_data, dtype = np.bool)
# mask[np.triu_indices_from(mask)] = True
```

```python
f, ax = plt.subplots(figsize = (11, 9))
```

```python
cmap = sns.diverging_palette(220, 10, as_cmap = True)
```

```python
sns.set(style = 'white')

labelsxy = ['A', 'B', 'C', 'D', 'E', 'F']

heatmap_corr = sns.heatmap(corr_matrix_data, mask = mask,
                           annot = True,
                           square = True,
                           cmap = 'Blues')
#                           xticklabels= labelsxy, 
#                           yticklabels= labelsxy)
```

```python
labelsxy = ['A', 'B', 'C', 'D', 'E', 'F']

heatmap_corr = sns.heatmap(corr_matrix_data, 
                           annot = True, 
                           robust = True,
                           xticklabels= labelsxy, yticklabels= labelsxy,)

plt.show()
```
